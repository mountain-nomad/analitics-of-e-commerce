# 5. Используя pandas, проведи когортный анализ пользователей. В период с января по декабрь выяви когорту с самым высоким retention на 3й месяц.(15 баллов)

```python
import pandas as pd

customers = pd.read_csv('olist_customers_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
items = pd.read_csv('olist_order_items_dataset.csv')

orders.order_purchase_timestamp = pd.to_datetime(orders.order_purchase_timestamp)
orders.order_delivered_customer_date = pd.to_datetime(orders.order_delivered_customer_date)
orders.order_estimated_delivery_date = pd.to_datetime(orders.order_estimated_delivery_date)
orders.order_approved_at = pd.to_datetime(orders.order_approved_at)
orders.order_delivered_carrier_date = pd.to_datetime(orders.order_delivered_carrier_date)
```

```python
# Объединение таблиц
merged_data = pd.merge(items, orders, on="order_id")
merged_data = pd.merge(merged_data, customers, on="customer_id")
```

```python
# преобразование дат в нужный формат
merged_data['order_purchase_timestamp'] = pd.to_datetime(merged_data['order_purchase_timestamp'])
```

```python
# Извлечение месяца первого заказа для каждого клиента
merged_data['order_month'] = merged_data['order_purchase_timestamp'].dt.to_period('M')
merged_data['cohort_month'] = merged_data.groupby('customer_unique_id')['order_month'].transform('min')
```

```python
# время жизни когорт в месяцах
merged_data['cohort_lifetime'] = (merged_data['order_month'].dt.year - merged_data['cohort_month'].dt.year) * 12 + (merged_data['order_month'].dt.month - merged_data['cohort_month'].dt.month) + 1
```

```python
# сводная таблица с количеством уникальных клиентов в каждой когорте и месяце
cohort_counts = merged_data.groupby(['cohort_month', 'cohort_lifetime']).agg({'customer_unique_id': 'nunique'}).reset_index()
```

```python
# общее количество клиентов в первом месяце каждой когорты
initial_customer_count = cohort_counts[cohort_counts['cohort_lifetime'] == 1][['cohort_month', 'customer_unique_id']]
```

```python
initial_customer_count = initial_customer_count.rename(columns={'customer_unique_id': 'cohort_customer_count'})
```

```python
cohort_counts = pd.merge(cohort_counts, initial_customer_count, on='cohort_month')
```

```python
# retention rate
cohort_counts['retention'] = cohort_counts['customer_unique_id'] / cohort_counts['cohort_customer_count']
```

```python
# Выберите когорту с самым высоким retention на 3-й месяц
highest_retention_cohort = cohort_counts[cohort_counts['cohort_lifetime'] == 3].sort_values(by='retention', ascending=False).iloc[0]

highest_retention_cohort
```

# 6. Часто для качественного анализа аудитории использую подходы, основанные на сегментации. Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию. В кластеризации можешь выбрать следующие метрики: R - время от последней покупки пользователя до текущей даты, F - суммарное количество покупок у пользователя за всё время, M - сумма покупок за всё время. Подробно опиши, как ты создавал кластеры. Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров. Пример такого описания: RFM-сегмент 132 (recency=1, frequency=3, monetary=2) имеет границы метрик recency от 130 до 500 дней, frequency от 2 до 5 заказов в неделю, monetary от 1780 до 3560 рублей в неделю. Описание подхода можно найти тут. (23 балла)

```python
import pandas as pd

# Загрузка данных
customers = pd.read_csv('olist_customers_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
items = pd.read_csv('olist_order_items_dataset.csv')

orders.order_purchase_timestamp = pd.to_datetime(orders.order_purchase_timestamp)
orders.order_delivered_customer_date = pd.to_datetime(orders.order_delivered_customer_date)
orders.order_estimated_delivery_date = pd.to_datetime(orders.order_estimated_delivery_date)
orders.order_approved_at = pd.to_datetime(orders.order_approved_at)
orders.order_delivered_carrier_date = pd.to_datetime(orders.order_delivered_carrier_date)
```

```python
# Объединение данных
merged_data = pd.merge(customers, orders, on="customer_id")
merged_data = pd.merge(merged_data, items, on="order_id")
```

```python
merged_data["order_purchase_timestamp"] = pd.to_datetime(merged_data["order_purchase_timestamp"])
merged_data["order_delivered_customer_date"] = pd.to_datetime(merged_data["order_delivered_customer_date"])
```

```python
# Рассчет Recency, Frequency и Monetary
current_date = pd.to_datetime("2023-11-13")
rfm_table = merged_data.groupby("customer_id").agg({
    'order_purchase_timestamp': lambda x: (current_date - x.max()).days,
    'order_id': 'count',
    'price': 'sum'
})

rfm_table.rename(columns={'order_purchase_timestamp': 'recency',
                          'order_id': 'frequency',
                          'price': 'monetary'}, inplace=True)
```

```python
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_table)
rfm_scaled = pd.DataFrame(rfm_scaled, index=rfm_table.index, columns=rfm_table.columns)
```

```python
# Определение числа кластеров (может потребоваться подобрать оптимальное значение)
kmeans = KMeans(n_clusters=3, random_state=42)
rfm_table['cluster'] = kmeans.fit_predict(rfm_scaled)
```

```python
cluster_summary = rfm_table.groupby('cluster').agg({
    'recency': ['min', 'max'],
    'frequency': ['min', 'max'],
    'monetary': ['min', 'max']
}).reset_index()

# Вывод интерпретации
for i, row in cluster_summary.iterrows():
    print(f"Cluster {int(row['cluster'])}:")
    print(f"Recency: {int(row['recency']['min'])} to {int(row['recency']['max'])} days")
    print(f"Frequency: {int(row['frequency']['min'])} to {int(row['frequency']['max'])} orders")
    print(f"Monetary: {int(row['monetary']['min'])} to {int(row['monetary']['max'])} monetary units")
    print("\n")
```
